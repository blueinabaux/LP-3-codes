```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
# from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import scale
```

```python
try:
    data = pd.read_csv('emails.csv')
except FileNotFoundError:
    print("Error: 'emails.csv' not found. Please download it from the link in your manual and place it in the working directory.")
```

```python
data
```

```python
df = data.copy()
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.drop(['Email No.'], axis=1, inplace=True)
X = df.drop(['Prediction'], axis=1)
```

```python
y = df['Prediction']
```

```python
# 4. Normalize the data (Scale features, a common practice before KNN and SVM)
X_scaled = scale(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
print("--- Data Preprocessing Complete ---")
print(f"Training data shape (X_train): {X_train.shape}")
print(f"Testing data shape (X_test): {X_test.shape}")
print("-" * 35)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=7)
knn_model.fit(X_train, y_train)
```

```python
y_pred_knn = knn_model.predict(X_test)
knn_accuracy = accuracy_score(y_test, y_pred_knn)
```

```python
print("## K-Nearest Neighbors (KNN) Performance")
print(f"KNN Accuracy Score: {knn_accuracy:.4f}")
print("\nConfusion Matrix (KNN):\n", confusion_matrix(y_test, y_pred_knn))
print("\nClassification Report (KNN):\n", classification_report(y_test, y_pred_knn))
print("-" * 35)
```

```python
svm_model = SVC(C=1, kernel='rbf', random_state=42) 
svm_model.fit(X_train, y_train)
```

```python
y_pred_svm = svm_model.predict(X_test)
svm_accuracy = accuracy_score(y_test, y_pred_svm)
```

```python
print("## Support Vector Machine (SVM) Performance")
print(f"SVM Accuracy Score: {svm_accuracy:.4f}")
print("\nConfusion Matrix (SVM):\n", confusion_matrix(y_test, y_pred_svm))
print("\nClassification Report (SVM):\n", classification_report(y_test, y_pred_svm))
print("-" * 35)
```

```python
print("## Model Performance Comparison")
print(f"KNN Classifier Accuracy: {knn_accuracy:.4f}")
print(f"SVM Classifier Accuracy: {svm_accuracy:.4f}")
```

```python
if svm_accuracy > knn_accuracy:
    print("\nObservation: The SVM Classifier performed better than the KNN Classifier for this dataset.")
elif knn_accuracy > svm_accuracy:
    print("\nObservation: The KNN Classifier performed better than the SVM Classifier for this dataset.")
else:
    print("\nObservation: Both classifiers achieved the same accuracy.")
```
