```python
def func(x):
    return (x+3)**2
def derivative(x):
    return 2*(x+3)
```

```python
gd = []
def gradient_descent(start, learning_rate, n_iterations):
    x = start
    for i in range(n_iterations):
        grad = derivative(x)
        x = x - learning_rate * grad
        gd.append(x)
    return x
        
```

```python
starting_x = 2
learning_rate = 0.1
n_iterations = 50

min_x = gradient_descent(starting_x, learning_rate, n_iterations)
print("The minimum value of x: ", min_x)
```

```python
import seaborn as sns
sns.scatterplot(gd)
```
